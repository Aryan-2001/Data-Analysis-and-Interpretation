## PEARSON CORRELATION ON GAPINDER DATA SET 

### IMPORTING LIBRARIES 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import scipy.stats as sp
```

### Importing Dataset 

```python
data = pd.read_csv("gapminder.csv")
data.head()
```

## Data Mangement

```python
data2 = data.iloc[:,[0,1,2,9,14]].copy()
```

```python
data2.columns = ["country","income","alcohol","life","employment"]
```

```python
def fun(a):
    
    for i in range(len(a)):
        if a[i]==" ":
            a[i] = np.nan
        else:
            a[i] = float(a[i])
            
    
```

```python
for i in data2.columns[1:]:
    fun(data2[i])
```

```python
def fun2(a):
    x = np.mean(a.dropna())
    for i in range(len(a)):
        if np.isnan(a[i]):
            a[i]=x
```

```python
for i in data2.columns[1:]:
    fun2(data2[i])
```

# INCOME VS ALCOHOL---

```python
plt.scatter(data2["income"],data2["alcohol"])
```

```python
sp.pearsonr(data2["income"],data2["alcohol"])
```

This shows r is very near 0 and p value is very low i.e so we can say that income is very weakly dependant on alcohol
