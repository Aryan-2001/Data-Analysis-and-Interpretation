# Handling missing values

```python
for i in range(len(data)):
    
    if data.iloc[i,1]==" ":
        data.iloc[i,1] = np.nan
    else:
        data.iloc[i,1] = float(data.iloc[i,1])
    
    
```

```python
data2 = data.iloc[ : ,[0,1,2,9,14]].copy()
```

```python
data2
```

```python
data2.columns = ["country","income","alcohol","life","employment"]
```

```python
data2.head()
```

```python
x = np.mean(data2["income"].dropna())
```

```python
data2["income"] = data2["income"].replace(to_replace = np.nan , value = x)
```

```python
data2
```

```python
y = np.mean(data2["alcohol"].dropna())
data2["alcohol"] = data2["alcohol"].replace(np.nan , y)
```

```python
data2
```

```python
z = np.mean(data2["life"].dropna())
data2["life"] = data2["life"].replace(np.nan,z)
```

```python
data2
```

```python
a = np.mean(data2["employment"].dropna())
data2["employment"] = data2["employment"].replace(np.nan , a)
data2
```
